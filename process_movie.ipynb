```python
def convert_movie(movie):
    # 定数定義
    ESC_KEY = 27     # Escキー
    INTERVAL= 33     # インターバル
    FRAME_RATE = 30  # fps

    WINDOW_ORG = "org"
    WINDOW_BACK = "back"
    WINDOW_DIFF = "diff"
    DIFF_BOOL='bool'

    diff_pixels = []

    # def movie_loader(movie):
    FILE_ORG = BASE_FILE + movie
    diff_pixels = []
    mov_org = cv2.VideoCapture(FILE_ORG)

    # 最初のフレーム読み込み
    has_next, i_frame = mov_org.read()
    count = 1

    # 背景フレーム
    back_frame = np.zeros_like(i_frame[55:, :], np.float32)

    # 変換処理ループ
    while has_next == True:
        if count % 500 == 0:
            print('finshed {}'.format(count))
        i_frame = i_frame[55:, :]
        # 入力画像を浮動小数点型に変換
        f_frame = i_frame.astype(np.float32)

        # 差分計算
        diff_frame = cv2.absdiff(f_frame, back_frame)

        # 背景の更新
        cv2.accumulateWeighted(f_frame, back_frame, 0.025)

        mask = diff_frame.copy()
        mask2 = diff_frame.copy()
        mask3 = diff_frame.copy()
        mask4 = diff_frame.copy()
        th = 30
        mask[mask < th] = 0
        mask[mask >= th] = 255
        th=50
        mask2[mask2 < th] = 0
        mask2[mask2 >= th] = 255
        th = 100
        mask3[mask3 < th] = 0
        mask3[mask3 >= th] = 255
        th=150
        mask4[mask4 < th] = 0
        mask4[mask4 >= th] = 255
#         cv2.imshow(DIFF_BOOL, mask2)

#         # Escキーで終了
#         key = cv2.waitKey(INTERVAL)
#         if key == ESC_KEY:
#             break

        # 次のフレーム読み込み
        has_next, i_frame = mov_org.read()
        count += 1
#         if count % 120:
#             diff_pixels.append([movie, count,diff_frame])
        diff_pixels.append([movie, count, len(mask[mask >= 1]), len(mask2[mask2 >= 1]),len(mask3[mask3 >= 1]), len(mask4[mask4 >= 1])])
    #         i_frame = i_frame[55:, :]
    with open(BASE_FILE + 'video_{}.pickle'.format(movie), 'wb') as f:
        pickle.dump(diff_pixels, f)

    # 終了処理
    cv2.destroyAllWindows()
    mov_org.release()
```
